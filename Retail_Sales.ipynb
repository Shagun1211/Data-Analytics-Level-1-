## Visualizations

```python
var = df.groupby('Gender')['Total Amount'].sum()
```

```python
var.plot(kind="bar", title="Total Sales grouped by Gender", xlabel="Gender", ylabel="Total Amount", figsize=(6,6))
plt.xticks(rotation=0)
plt.yticks(rotation=0)
```

```python
var2 = df.groupby('Product Category')['Total Amount'].sum()
```

```python
var2.plot(kind="bar", title="Total Sales grouped by Product Category", xlabel="Product Category", ylabel="Total Amount", figsize=(7,6), color="pink")
plt.xticks(rotation=0)
plt.yticks(rotation=0)
```

```python
df['Age'].unique()
```

```python
total_sales_below_40 = df[df['Age'] < 40]['Total Amount'].sum()
total_sales_above_40 = df[df['Age'] > 40]['Total Amount'].sum()

mean_values = pd.Series({
    'Age < 40' : total_sales_below_40 ,
    'Age > 40' : total_sales_above_40
})
mean_values.plot(kind='bar', title='Total Sales by Age group', xlabel='Age', ylabel='Total Amount', color='purple')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
```

## Time Series Analysis

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

df.set_index('Date', inplace=True)
```

```python
monthly_sales = df['Total Amount'].resample('M').sum()
weekly_sales = df['Total Amount'].resample('W').sum()
```

```python
weekly_sales.plot(title="Weekly Sales Trend", figsize=(10,5))
```

```python
monthly_sales.plot(title="Monthly Sales Trend", figsize=(10,5))
```

```python
sns.histplot(df['Age'], bins=10, kde=True)
```

```python
heatmap_data = df.groupby(['Month', 'Product Category'])['Total Amount'].sum().unstack()
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap='YlGnBu', linewidths=0.5)

plt.title('Monthly Sales by Product Category', fontsize=14)
plt.xlabel('Product Category')
plt.ylabel('Month')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Recommendations

## 1. Product Strategy

Top-Selling Categories:
Electronics and Clothing are the top-performing categories with total sales of ₹156,905 and ₹155,580 respectively.
Focus future marketing, inventory stocking, and bundle deals on these categories.

Underperforming Category:
Beauty products contribute the least (₹143,515). Analyze pricing, display positioning, or promotional gaps to boost performance.


## 2.  Customer Demographics & Targeting

Gender Balance:
Female customers slightly outnumber males (51% vs. 49%).
Launch women-focused campaigns or products, especially in categories like Clothing and Beauty.

Age Insights:
Majority of customers are aged 29 to 53, with a mean age of ~41.
Target this age group with loyalty programs and personalized offers.

## 3.  Transaction Insights

Quantity & Pricing:
Average quantity per transaction: ~2.5 items
Most products are priced around ₹50–₹300.

Promote combo offers or “Buy 2, get 1” campaigns to match the buying pattern and encourage higher cart values.


## 4.  Category Trends Over Time (Heatmap Insight)

Sales of Electronics peak during festive months (e.g., October & December). Use this data for predictive inventory planning.
Clothing shows consistent performance — ideal for monthly subscriptions or style-box offers.
